```python
import warnings
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
warnings.filterwarnings('ignore')
```

```python
algorithms = ["baseline", "omp", "atomic", "reduction"] #, "lockStriping"]
machines = ["hannespc", "ironmaiden"]
```

```python
filenames = []
for algorithm in algorithms:
    for machine in machines:
        filenames.append(f"{machine}_{algorithm}.csv")
```

```python
with open('results.csv', 'w') as outfile:
    for fname in filenames:
        with open(fname) as infile:
            for line in infile:
                outfile.write(f"{fname[:fname.find('_')]},{line}")
```

```python
columns = ["Machine", "Type", "Dataset", "M", "Filter size", "Filter #k", "False pos", "False pos rate", "Expected false pos rate", "Insert time", "Query time", "Total time", "Threads"]
```

```python
results_df = pd.read_csv("results.csv", header = None)
results_df.columns = columns
```

```python
len(results_df)
```

```python
nuovo_df_probabilmente = pd.DataFrame(columns=columns)
grouped = results_df.groupby(['Machine', 'Type', 'Dataset', 'M', 'Threads'])
for name, group in grouped:
    group = group.sort_values("Total time")
    group = group.drop(group.tail(1).index)
    group = group.drop(group.head(1).index)
    #print(group.mean())
    media_fuori = group.mean(numeric_only=True)
    #print(name[0])
    media_fuori["Machine"] = name[0]
    media_fuori["Type"] = name[1]
    media_fuori["Dataset"] = name[2]
    nuovo_df_probabilmente = nuovo_df_probabilmente.append(media_fuori, ignore_index=True)#.reset_index())
```

```python
g = sns.FacetGrid(nuovo_df_probabilmente, col="M", row="Dataset", height=5, aspect=2)

g.map(sns.barplot, "Threads", "Insert time", "Type", palette=sns.color_palette("tab10")).add_legend()
```

```python
for name, group in nuovo_df_probabilmente.groupby(['Dataset', 'M']):
    g = sns.FacetGrid(data=group, col="Machine", row="Type", height=5, aspect=2, margin_titles=True)
    g.fig.subplots_adjust(top=0.9) # adjust the Figure in rp
    g.fig.suptitle('Dataset: ' + name[0] + "     Size: " + str(int(name[1]/1000000)) + "M")
    g.map(sns.barplot, "Threads", "Insert time", palette=sns.color_palette("tab10"))
```

```python
speed_df = pd.DataFrame(columns=columns + ['speedup'])
```

```python
for name, group in nuovo_df_probabilmente.groupby(["Machine", "Type", "Dataset", "M"]):
    one_thread = group[group['Threads'] == 1]['Total time']
    #print(f"ONE_THREAD {one_thread}")
    #print(group['Total time'])
    group['speedup'] = group['Total time'].apply(lambda x: one_thread / x)
    speed_df = speed_df.append(group, sort=False)
```
