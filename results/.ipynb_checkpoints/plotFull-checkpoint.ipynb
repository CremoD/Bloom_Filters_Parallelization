```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
columns = ["Type", "Dataset", "M", "Filter size", "Filter #k", "False pos", "False pos rate", "Expected false pos rate", "Insert time", "Query time", "Total time", "Threads"]
```

```python
baseline_df = pd.read_csv(f"baseline{cluster}.csv", header = None)
omp_df = pd.read_csv(f"omp{cluster}.csv", header = None)
atomic_df = pd.read_csv(f"atomic{cluster}.csv", header = None)
reduction_df = pd.read_csv(f"reduction{cluster}.csv", header = None)
```

```python
baseline_df.columns = columns
omp_df.columns = columns
atomic_df.columns = columns
reduction_df.columns = columns
```

```python
df_list = []
df_list.append(baseline_df)
df_list.append(omp_df)
df_list.append(atomic_df)
df_list.append(reduction_df)
```

```python
#if is_cluster:
#    lockStriping_df = pd.read_csv(f"lockStriping{cluster}.csv", header = None)
#    lockStriping_df.columns = columns
#    df_list.append(lockStriping_df)

#    lockStripingPL_df = pd.read_csv(f"lockStripingPL{cluster}.csv", header = None)
#    lockStripingPL_df.columns = columns
#    df_list.append(lockStripingPL_df)
```

```python
for i in range(len(df_list)):
    res = df_list[i].sort_values('Total time')
    res = res.drop(res.groupby(['Type', 'Dataset', 'M', 'Threads']).tail(1).index)
    df_list[i] = res.drop(res.groupby(['Type', 'Dataset', 'M', 'Threads']).head(1).index)
    df_list[i] = df_list[i].groupby(['Type', 'Dataset', 'M', 'Threads']).mean().reset_index()
```

```python
all_in = pd.DataFrame(columns=columns)
all_in = df_list[0]
for el in range(1,len(df_list)):
    all_in = pd.concat([all_in, df_list[el]])
```

# COMPARISON OF DIFFERENT APPROACHES

```python
for key, group in all_in.groupby("Dataset"):
    for i in group['M'].unique():
        df_filtered = group[group['M']== i]
        df_filtered.pivot("Threads", "Type", "Query time").plot(kind='bar', title=key + " - " + str(i), rot=0)
        #df_filtered.plot(x="Threads", y = ["Total time"], kind = "bar", title=key + " - 10^" + str(len(str(i)) - 1), rot=0)
        plt.ylabel("Milliseconds")
        plt.legend(loc='upper left')
```

```python
sns.set(font_scale=2.5)
```

```python
g = sns.FacetGrid(all_in, col="M", row="Dataset", height=5, aspect=2)

g.map(sns.barplot, "Threads", "Insert time", "Type", palette=sns.color_palette("tab10")).add_legend()
```

```python
g = sns.FacetGrid(all_in, col="M", row="Dataset", height=5, aspect=2)

g.map(sns.barplot, "Threads", "Query time", "Type", palette=sns.color_palette("tab10")).add_legend()
```

```python
speed_df = pd.DataFrame(columns=columns + ['speedup'])
```

```python
for name, group in all_in.groupby(["Type", "Dataset", "M"]):
    one_thread = group[group['Threads'] == 1]['Total time']
    #print(f"ONE_THREAD {one_thread}")
    #print(group['Total time'])
    group['speedup'] = group['Total time'].apply(lambda x: one_thread / x)
    speed_df = speed_df.append(group)
```

# SPEEDUP

```python
for key, group in speed_df.groupby("Dataset"):
    for i in group['M'].unique():
        df_filtered = group[group['M']== i]
        df_filtered.pivot("Threads", "Type", "speedup").plot(kind='bar', title=key, rot=0)
        #df_filtered.plot(x="Threads", y = ["Total time"], kind = "bar", title=key + " - 10^" + str(len(str(i)) - 1), rot=0)
        plt.ylabel("Speedup")
        plt.legend(loc='upper left')
```

```python
g = sns.FacetGrid(speed_df, col="M", row="Dataset", height=5, aspect=2)

g.map(sns.barplot, "Threads", "speedup", "Type", palette=sns.color_palette("tab10")).add_legend()
```

# CHUNK SIZE

```python
columns = ["Type", "Dataset", "M", "Filter size", "Filter #k", "False pos", "False pos rate", "Expected false pos rate", "Insert time", "Query time", "Total time", "Threads", "Chunk size"]

static_df = pd.read_csv("chunk_static.csv", header = None)
dynamic_df = pd.read_csv("chunk_dynamic.csv", header = None)
#openmp_atomic_df = pd.read_csv("openmp_atomic.csv", header = None)

static_df.columns = columns
dynamic_df.columns = columns
#openmp_atomic_df.columns = columns

static_df = static_df.sort_values(['Dataset', 'M'], ascending=[True, True])
dynamic_df = dynamic_df.sort_values(['Dataset', 'M'], ascending=[True, True])
```

```python
static_df.plot(kind="bar", title="static, dataset = 10^8", x="Chunk size", y="Insert time", color="#009933")
plt.ylabel("Milliseconds")
plt.tight_layout()
```

```python
dynamic_df.plot(kind="bar", title="dynamic, dataset = 10^8", x="Chunk size", y="Insert time", color="#cc0000")
plt.ylabel("Milliseconds")
plt.tight_layout()
```
