```python
from matplotlib import pyplot as plt
import pandas as pd
from matplotlib.backends.backend_pdf import PdfPages
```

```python
columns = ["Type", "Dataset", "M", "Filter size", "Filter #k", "False pos", "False pos rate", "Expected false pos rate", "Insert time", "Query time", "Total time", "Threads"]
```

```python
string_df = pd.read_csv("string.csv", header = None)
openmp_df = pd.read_csv("openmp_threads.csv", header = None)
#openmp_atomic_df = pd.read_csv("openmp_atomic.csv", header = None)
```

```python
string_df.columns = columns
openmp_df.columns = columns
#openmp_atomic_df = columns
```

```python
string_df = string_df.sort_values(['Dataset', 'M'], ascending=[True, True])
openmp_df = openmp_df.sort_values(['Dataset', 'M'], ascending=[True, True])
```

```python
df = pd.concat([string_df, openmp_df])
```

## OMP Baseline Threads comparison

```python
pdf = PdfPages(f"omp_baseline.pdf")

firstPage = plt.figure()
firstPage.clf()
txt = 'OpenMP Version: 4.5.0\nOS: Windows 10\nCPU: Intel Core I7-10750H\nPhysical cores: 6\nBase freq. CPU: 2.6GHZ'
firstPage.text(0.5,0.5,txt, transform=firstPage.transFigure, size=24, ha="center")
pdf.savefig()


for title, group in df.groupby("Dataset"):
    #group.pivot("M", "Type", "Total time").plot(kind='bar', title=title, rot=0)
    for i in group['M'].unique():
        df_filtered = group[openmp_df['M']== i]
        df_filtered.plot(x="Threads", y = ["Insert time", "Query time"], kind = "bar", title=title + " - 10^" + str(len(str(i)) - 1), rot=0)
        plt.ylabel("Milliseconds")
        plt.legend(loc='upper left')
    pdf.savefig()
pdf.close()
```
